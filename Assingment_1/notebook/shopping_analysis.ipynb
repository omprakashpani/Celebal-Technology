# Step 7: Insights and Conclusion

## Dataset Overview

- The dataset contains **1000 products** and **24 columns** related to product details, pricing, ratings, discounts, seller information, and categories.
- Products belong to multiple categories such as backpacks, watches, bath robes, yoga mats, and more.
- Missing values were identified in columns such as `discount`, `seller_name`, `videos`, `seller_information`, `what_customers_said`, and `variations`, and were handled during data cleaning.

---

## Insights from Original Columns

### Product and Category Information

- The dataset contains products from various categories, indicating a diverse product catalog.
- Some categories have a significantly higher number of products, suggesting greater market presence and customer demand.

### Rating (`rating`)

- Product ratings range from **0 to 5**.
- The average rating is approximately **3.62**, indicating generally positive customer feedback.
- Most products are rated between **3.5 and 4.5 stars**.

### Ratings Count (`ratings_count`)

- Customer engagement varies considerably across products.
- While many products have only a few ratings, some products have received thousands of ratings.
- Products with higher review counts tend to have greater visibility and customer trust.

### Initial Price (`initial_price`)

- Product prices vary widely across categories.
- The median price is around **₹1999**, while premium products can exceed **₹20,000**.
- This indicates the presence of both budget and luxury products.

### Discount (`discount`)

- Most products offer discounts.
- The average discount is approximately **53%**, showing that promotional pricing is a major sales strategy.

### Final Price (`final_price`)

- Final prices are generally lower than initial prices due to discounts.
- Final price represents the actual amount paid by customers after applying offers.

### Seller Information

- Multiple sellers contribute products to the marketplace.
- Missing seller information was handled to maintain dataset consistency.

---

## Insights from Derived Columns

### Price Difference (`price_difference`)

```text
price_difference = initial_price - final_price
```

- This feature measures the amount reduced from the original price.
- Larger values indicate heavily discounted products.
- It helps identify products offering significant savings to customers.

### Popularity (`popularity`)

```text
popularity = rating × ratings_count
```

- Popularity combines customer satisfaction and customer engagement into a single metric.
- Products with high ratings and many reviews achieve the highest popularity scores.
- This metric is useful for identifying top-performing products.

---

## Visualization Findings

### Rating Distribution

- Most products are concentrated around ratings of **4 stars**.
- Very few products have extremely low ratings.
- This indicates generally positive customer experiences.

### Price Distribution

- Most products fall within a moderate price range.
- A small number of high-priced products create price outliers.

### Category Analysis

- Certain categories contain significantly more products than others.
- These categories may represent high-demand market segments.

### Rating vs Popularity Analysis

- Products with higher ratings generally show higher popularity.
- However, popularity also depends on the number of customer reviews.
- A highly rated product with very few reviews may be less popular than a slightly lower-rated product with many reviews.

---

# Conclusion

- The dataset provides valuable insights into product pricing, customer ratings, discounts, and category performance.
- Data cleaning successfully handled missing values and ensured consistency across the dataset.
- Feature engineering enabled deeper analysis through the creation of **Price Difference** and **Popularity** metrics.
- The analysis shows that discounts significantly influence final prices and customer purchasing decisions.
- Products with strong ratings and large review counts emerge as the most popular products.
- Overall, exploratory data analysis helped uncover meaningful patterns in customer engagement, pricing strategies, and product performance.
